```python
import cv2
import numpy as np
import os
import time
import HandTracking as htm
```

```python
folderPath = "Header"
myList = os.listdir(folderPath)
print(myList)
overlayList = []
for imPath in myList:
    image = cv2.imread(f'{folderPath}/{imPath}')
    overlayList.append(image)
print(len(overlayList))
header = overlayList[0]
drawColor = (255, 255, 255)
brushThckness = 10
eraserThickness= 40

cap = cv2.VideoCapture(0)
cap.set(3, 1500)
cap.set(4, 720)

detector = htm.handDetector(detectionCon=0.85)
xp,yp = 0,0
imgCanvas = np.zeros((450,640,3), np.uint8)

while True:
    #Import Image
    success, img = cap.read()
    img = cv2.flip(img, 1) #Flipping the image to actusl sides
    
    #To detect Hand
    img = detector.findHands(img)
    lmList = detector.findPosition(img, draw=False)
    if len(lmList)!=0:
        #print(lmList)
            # tip of index and middle fingers
        x1, y1 = lmList[8][1:]
        x2, y2 = lmList[12][1:]
        
        
    
        #To check which finger is Up
        fingers = detector.fingersUp()
        # print(fingers)
    
        #Selection Mode: Two Fingers up
        if fingers[1] and fingers[2]:
            print("Selction Mode")
            #checking for click
            if y1 <100:
                if 100<x1<150:
                    header = overlayList[0]
                    drawColor= (203, 124, 252)
                elif 200<x1<250:
                    header = overlayList[1]
                    drawColor = (255, 122, 202)
                elif 380<x1<420:
                    header = overlayList[2]
                    #***drawColor = (138, 151, 255)
                elif 550<x1<600:
                    header = overlayList[3]
                    drawColor = (0,0,0)
            cv2.rectangle(img, (x1,y1-10), (x2,y2+10), drawColor, cv2.FILLED)
                    
            
        
        #Drawing Mode: Index Finger up
        if fingers[1] and fingers[2] == False:
            cv2.circle(img, (x1,y1), 10, drawColor, cv2.FILLED)
            print("Drawing Mode")
            if xp==0 and yp==0:
                xp, yp = x1, y1
                
            if drawColor ==(0,0,0):
                cv2.line(img, (xp, yp),(x1,y1), drawColor, eraserThickness)
                cv2.line(imgCanvas, (xp, yp),(x1,y1), drawColor, eraserThickness)
            
            else:
                cv2.line(img, (xp, yp),(x1,y1), drawColor, brushThckness)
                cv2.line(imgCanvas, (xp, yp),(x1,y1), drawColor, brushThckness)
            xp, yp = x1, y1
    
    
    
    #Setting the header image
    img[0:100, 0:640, :] = header
    #***img = cv2.addWeighted(img, 0.5, imgCanvas,0.5,0)
    cv2.imshow("Image", img)
    cv2.imshow("Canvas", imgCanvas)
    cv2.waitKey(1)    
    


    
```
